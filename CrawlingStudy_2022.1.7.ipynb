```python
import requests
```

```python
req = requests.get('https://tv.naver.com/r/')
raw = req.text

print(raw)

#requests: 외부 라이브러리&웹페이지에 요청할 수 있는 기능들을 모아놓은 클래스
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
html = BeautifulSoup(raw, 'html.parser')
```

```python
infos = html.select('div.cds')
```

```python
clip1 = infos[0]
clip1_title = clip1.select_one('dt.title tooltip')
print(clip1_title)
```

* select_one() 함수는 select() 함수와 마찬가지로 HTML 요소 내에서 매개변수로 전달된 선택자를 갖는 요소를 찾아줌. 
* 다른 점은 여러 개가 존재할 수 있는 요소도 맨 처음 나오는 하나만 돌려준다는 점
* 따라서 클립 정보 리스트와 같이 여러 개를 가져와야 할 경우는 select() 함수, 클립 내의 제목같이 하나만 존재하는 값을 가져올때는 select_one() 함수를 쓰는 것이 좋습니다.

```python
print(clip1.select_one('tooltip').text)
print(clip1.select_one('dd.chn > a').text)
print(clip1.select_one('span.hit').text)
print(clip1.select_one('span.like').text)
#clip1이 div.cds 선택자를 갖는 HTML이기 때문에, 서브 선택자 경로들은 div.cds 의 자손부터 작성하면 됩니다.
```

```python
#for문 활용
for info in infos:
    title = info.select_one('tooltip').text
    chn = info.select_one('dd.chn > a').text
    hit = info.select_one('span.hit').text
    like = info.select_one('span.like').text

    print(chn, '/', title, '/', hit, '/', like)
```

```python
#challenge1
#1~3위 클립의 클립명, 채널명, 재생 수, 조회 수를 출력
#형식은 클립명 / 채널명 / 재생 수 / 조회 수 
infos_2 = html.select('div.info')
clip2 = infos_2[0]

for info in infos_2:
    title = info.select_one('strong.tit').text
    chn = info.select_one('dd.chn > a').text #a.ch도 가능
    hit = info.select_one('span.hit').text
    like = info.select_one('span.like').text

    print(chn, '/', title, '/', hit, '/', like)
```

```python
#모범 답안
#infos에 대한 정의에서 차이 있으나 for문에서 같은 것을 지정했기 때문에 같은 결과
infos = html.select('ul.rolling > li')

for info in infos:
    title = info.select_one('strong.tit > span').text
    chn = info.select_one('a.ch').text
    hit = info.select_one('span.hit').text
    like = info.select_one('span.like').text

    print(chn, '/', title, '/', hit, '/', like)
```

```python
#다른페이지 적용
import requests
from bs4 import BeautifulSoup

raw = requests.get('https://search.naver.com/search.naver?&where=news&query=%EB%8C%80%EC%84%A0', headers={'User-Agent': 'Mozilla/5.0'}).text
html = BeautifulSoup(raw, 'html.parser')

articles = html.select('.list_news > li')

for article in articles:
    journal = article.select_one('a.info.press').text
    title = article.select_one('a.news_tit').text
    
    print(journal,'/',title)
```

```python
#여러 페이지 수집
for page in range(10):
    raw = requests.get('https://search.naver.com/search.naver?&where=news&query=대선&start=' + str(page * 10 + 1), headers={'User-Agent': 'Mozilla/5.0'}).text
    html = BeautifulSoup(raw, 'html.parser')
    articles = html.select('.list_news > li')

    for article in articles:
        journal = article.select_one('a.info.press').text
        title = article.select_one('a.news_tit').text

        print(journal, title)
```
