```python
#ë‹¤ë¥¸í˜ì´ì§€ ì ìš©
import requests
from bs4 import BeautifulSoup

raw = requests.get('https://search.naver.com/search.naver?&where=news&query=%EB%8C%80%EC%84%A0', headers={'User-Agent': 'Mozilla/5.0'}).text
html = BeautifulSoup(raw, 'html.parser')

articles = html.select('.list_news > li')

for article in articles:
    journal = article.select_one('a.info.press').text
    title = article.select_one('a.news_tit').text
    
    print(journal,'/',title)
```
